## Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from sklearn.cluster import KMeans
```

## Importing Dataset

```python
customers = pd.read_csv('Mall_Customers.csv')
df = customers.copy()
df.head()
```

```python
X = df[['Annual Income (k$)','Spending Score (1-100)']].values
```

## Using the Elbow Method to Find Cluster Number

```python
wcss = [] # Within-Cluster-Sum-of-Squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (5,3))
plt.plot(range(1, 11), wcss)
plt.xticks(range(1,11))
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Create KMeans Cluster

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
clusters = kmeans.fit_predict(X)

df['Cluster'] = clusters
```

## Plotting the Clusters

```python
plt.figure(figsize = (10,5))
plt.scatter(X[clusters == 0, 0], X[clusters == 0, 1], s = 75, c = 'red', label = 'Cluster 1')
plt.scatter(X[clusters == 1, 0], X[clusters == 1, 1], s = 75, c = 'blue', label = 'Cluster 2')
plt.scatter(X[clusters == 2, 0], X[clusters == 2, 1], s = 75, c = 'green', label = 'Cluster 3')
plt.scatter(X[clusters == 3, 0], X[clusters == 3, 1], s = 75, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[clusters == 4, 0], X[clusters == 4, 1], s = 75, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 200, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

## Predict a Single Value

```python
try:
    annual_income = int(input("Annual Income (k$)\t: "))
    spending_score = int(input("Spending Score (1-100)\t: "))

    if(spending_score<0 or spending_score > 100):
        print("ERROR!\tSpending score value must be between 0 and 100!")
    else:
        prediction = kmeans.predict([[annual_income,spending_score]])
        print(f"Predicted cluster is \t: {prediction[0]+1}")
    
except ValueError:
    print("ERROR!\tInvalid value!")
```
